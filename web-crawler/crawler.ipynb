```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://resources.workable.com/job-descriptions/')
print(res.text)
```

```python
soup = BeautifulSoup(res.text, "html.parser")
all_links = soup.find_all('a')
# target_element = soup.select_one("#main > div.section-padding > div > div > div.col-md-9.order-md-3 > div > section:nth-child(1) > ul > li:nth-child(3) > a")
```

```python
title_list = []
link_list = []

for link in all_links:
        href_value = link.get('href')
        title_value = link.get('title')
        if title_value != None:
                title_list.append(title_value)
                link_list.append(href_value)
```

```python
print(len(title_list))
print(len(link_list))
```

```python
data = {'JobTitle':title_list, 'link': link_list}
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(data)
```

```python
df['Job_Function'] = df['JobTitle'].apply(lambda x: x if x.endswith("job descriptions") else None).ffill()
```

```python
df = df[~df['JobTitle'].str.endswith("job descriptions")]
df.reset_index(drop=True, inplace = True)
```

```python
df['Job_Function'] = df['Job_Function'].str.replace("job descriptions","")
```

```python
df['Job_Function'] = df['Job_Function'].str.strip()
```

```python
df['JobTitle'] = df['JobTitle'].str.replace("job\xa0description","")
```

```python
from bs4 import BeautifulSoup
import json
import re
import pandas as pd
import requests

# Dictionary to store results for each job function
all_results = {}

for i in range(len(df)):
    job_function = df['Job_Function'][i]
    job_title = df['JobTitle'][i]
    url = df['link'][i]  # Replace with your actual column name for the URL

    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
        response = requests.get(url, headers=headers)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract information from <h2> tags
        results = []
        print(f'run to -----{i}')

        for h2_tag in soup.find_all('h2'):
            try:
                h2_contents_text = h2_tag.text.strip()
            except (IndexError, AttributeError):
                h2_contents_text = None
            text_between_h2 = ''
            if h2_contents_text and not h2_contents_text.startswith('I ') and h2_contents_text != 'Frequently asked questions':
                for sibling in h2_tag.find_all_next():
                    if sibling.name == 'h2':
                        break
                    elif sibling.name == 'div' and 'class' in sibling.attrs and 'col' in sibling['class']:
                        break
                    elif sibling.name == 'div':
                        break
                    elif sibling.name == 'a' and 'class' in sibling.attrs and 'article-banner' in sibling['class']:
                        break
                    sibling_text = re.sub(r'<ul>\n<li>|<li style=\"font-weight: 400;\">|<ul>\n<li style=\"font-weight: 400;\">|<span style=\"font-weight: 400;\">|<ul>\n<li style=\"font-weight: 400;\">|</li>\n</ul>|<span style=\"font-weight: 400;\">|<li style="font-weight: 400">|<span style="font-weight: 400">|<ul>\n<li style="font-weight: 400"><span style="font-weight: 400">|</span></li>\n<li style="font-weight: 400">|\xa0</span>|</span>|</span></li>\n</ul><li style="font-weight: 400">', '', str(sibling))
                    text_between_h2 += sibling_text
                    escaped_jobtitle = re.escape(job_title)
                    h2_contents_text = re.sub(f'(?i){escaped_jobtitle}', 'job', h2_contents_text)
                    h2_contents_text = re.sub(r'an', 'a', h2_contents_text)

                results.append({
                    h2_contents_text.strip(): text_between_h2.strip()
                })

        # Create a new dictionary for the current job position
        job_position_dict = {"Job_Title": job_title, "JD": {k: v for d in results for k, v in d.items()}}

        # Append the result to the list for the current job function
        if job_function not in all_results:
            all_results[job_function] = {"Job_Function": job_function, "Job_Position": []}
        all_results[job_function]["Job_Position"].append(job_position_dict)
    except requests.exceptions.RequestException as e:
        print(f"Error for URL {url}: {e}")

# Convert the final dictionary to JSON format
json_result = json.dumps(list(all_results.values()), indent=2)

# Save the JSON data to a file
output_filename = "all_job_descriptions_ver2.json"
with open(output_filename, 'w') as json_file:
    json_file.write(json_result)

print(f"All job descriptions saved to {output_filename}")
```
